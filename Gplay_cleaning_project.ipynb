```python
def clean_size(col):
    if "k" in col:
        col=col.replace("k","")
        return float(col)/1024
    elif "M"  in col:
        col=col.replace("M","")
        return float(col)
    elif "Varies with device" in col:
        return np.nan
    else:
        return float(col)
```

```python
df["Size"]=df["Size"].map(clean_size)
```
